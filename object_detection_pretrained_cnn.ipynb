###*Import TensorFlow's Keras API, Image packages*

```
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input, decode_predictions
```

```
from tensorflow.keras.preprocessing.image import load_img, img_to_array
```

###*Download the vgg16 model*

```
model = VGG16()
```

```
model.summary()
```

###*Import matplotlib library for plotting and os library for directory operations*

```
import os
import matplotlib.pyplot as plt
```

###*Load the model, loop through each picture, predict the output*

```
lstimg = []
lstlbl = []

imgFolder="Object-Detection_single-object/Img"

for file in os.listdir(imgFolder):
  
  print(file)
  full_path = imgFolder + '/' + file
  
  image = load_img(full_path,target_size=(224,224,3))
  img = image
  image = img_to_array(image)
  image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))

  image = preprocess_input(image)
  y_pred = model.predict(image)

  label=decode_predictions(y_pred, top=1)
  lstlbl.append(label)
  lstimg.append(img)
  
```
